# Python Lists

## What is a List?
A list stores multiple items in a single variable, created with square brackets.

```python
mylist = ['apple', 'banana', 'cherry']
```

<b>Key properties:</b>
- Ordered : items have a fixed index; new items go to the end
- Changeable (Mutable) : you can add, change, remove items after creation
- Allow duplicates : same value can appear multiple times
- Any data type : strings, ints, booleans, mixed types all valid

```python
list1 = ['abc', 34, True, 40, 'male']
print(len(list1))
print(type(list1))
```

<b>Alternative constructor:</b>

```python
thislist = list(("apple", "banana", "cherry")) # note double brackets
```

## Accessing Items

```python
thislist = ["apple", "banana", "cherry"]

thislist[0]   # "apple"  (first)
thislist[-1]  # "cherry" (last)
thislist[-2]  # "banana" (second from end)
```

### Slicing - list[start:end] (start included, end excluded)

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "melon", "mango"]

thislist[2:5]   # ["cherry", "orange", "kiwi"]
thislist[:4]    # first 4 items
thislist[2:]    # from index 2 to end
thislist[-4:-1] # ["orange", "kiwi", "melon"]
```

### Check existence

```python
if "apple" in thislist:
    print("Found!")
print(thislist.count('apple'))
```

## Change List items

### Change Item Value
To change the value of a specific item, refer to the index value.

```python
thislist = ["apple", "banana", "cherry"]
thislist[1] = "blackcurrant"
print(thislist)
```

### Change a Range of Item Values
To change the value of items withing a specific range, define a list with the new values, and refer to the range of index numbers where you want to insert the new values.

```python
thislist = ["apple", "banana", "cherry", "orange", "kiwi", "mango"]
thislist[1:3] = ["blackcurrant", "watermelon"]
print(thislist)
```

If you insert <b>more</b> items than you replace, the new items will be inserted where you specified, and the remaining items will move accordingly.

```python
thislist = ["apple", "banana", "cherry"]
thislist[1:2] = ["blackcurrant", "watermelon"]
print(thislist)
```

If you insert <b>less</b> items than you replace, the new items will be inserted where you specified, and the remaining items will move accordingly.

```python
thislist = ["apple", "banana", "cherry"]
thislist[1:3] = ["watermelon"]
print(thislist)
```

## Adding Items

```python
thislist.append("orange")             # add to end
thislist.insert(2, "watermelon")      # insert at index 2
thislist.extend(["mango", "pineapple"])  # add multiple from another list
thislist.extend(("kiwi", "papaya"))   # works with any iterable (tuple too)
```

## Removing Items

```python
thislist.remove("banana")   # removes first occurrence of value
thislist.pop(1)              # removes item at index 1 (returns it)
thislist.pop()               # removes last item
del thislist[0]              # deletes item at index 0
del thislist                 # deletes the entire list
thislist.clear()             # empties the list (list still exists)
```

## Looping

```python
print(thislist)
# for loop
for x in thislist:
    print(x)
print()
# loop by index
for i in range(len(thislist)):
    print(thislist[i])
print()
# while loop
i = 0
while i < len(thislist):
    print(thislist[i])
    i += 1
print()
# list comprehension (also a loop)
[print(x) for x in thislist]
```

## List Comprehension
Concise way to create a new list from an existing one.

```python
# syntax: [expression for item in iterable if condition]
fruits = ['apple', 'apricot', 'orange', 'grapes', 'guava', 'banana']
newlist = [x for x in fruits if "a" in x]
newlist = [x for x in range(10) if x < 5]
newlist = [x.upper() for x in fruits]
newlist = ["mango" if x == "banana" else x for x in fruits]
print(newlist)
```

if/else at the start is a ternary expression. Like regular expression, it sits at the front of the expression.

## Sorting

```python
thislist.sort()                        # ascending alphabetical / numeric
thislist.sort(reverse=True)            # descending
thislist.sort(key=str.lower)           # case-insensitive sort
thislist.reverse()                     # just reverses the order (not sort)

# custom sort with key function
def myfunc(n):
    return abs(n - 50)

thislist.sort(key=myfunc)
```

### key parameter
It accepts a function that is called on each item before comparing. The sort is then based on the return value of that function, not the item itself.

```python
thislist.sort(key=str.lower)       # built-in function
thislist.sort(key=len)             # sort by length
thislist.sort(key=myfunc)          # custom function

def myfunc(n):
    return abs(n - 50)

thislist = [100, 50, 65, 82, 23]
thislist.sort(key=myfunc)
# sorts by "distance from 50": [50, 65, 23, 82, 100]
```

## Copying a List
Don't use list2 = list1 - that just creates a reference, not a copy.

```python
mylist = thislist.copy()         # method 1
mylist = list(thislist)          # method 2
mylist = thislist[:]             # method 3 (slice)
```

## Joining Lists

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list3 = list1 + list2            # using + operator
list1.extend(list2)              # extend in-place
for x in list2:
    list1.append(x)              # manual loop
```

## List Unpacking

```python
fruits = ["apple", "banana", "cherry"]

a, b, c = fruits
print(a)  # apple
print(b)  # banana
print(c)  # cherry
```

Number of variables must match number of items - otherwise you get <mark>ValueError</mark>.

### <mark>*</mark> to collect the rest
If you don't want to unpack every item, use * to collect the leftovers into a list.

```python
fruits = ["apple", "banana", "cherry", "mango", "kiwi"]

a, *b, c = fruits
# a = "apple"
# b = ["banana", "cherry", "mango"]  ← middle items collected
# c = "kiwi"

first, *rest = fruits
# first = "apple"
# rest  = ["banana", "cherry", "mango", "kiwi"]

*start, last = fruits
# start = ["apple", "banana", "cherry", "mango"]
# last  = "kiwi"
```

Only <b>one *</b> variable is allowed per unpacking statement.

```python
a, b = 1, 2
a, b = b, a   # swap — uses unpacking under the hood
print(a, b)   # 2 1
```

## List Methods

- append() : Adds an element at the end of the list
- clear() : Removes all the elements from the list
- copy() : Returns a copy of the list
- count() : Returns the number of elements with the specified value
- extend() : Add the elements of a list (or any iterable), to the end of the current list
- index() : Returns the index of the first element with the specified value
- insert() : Adds an element at the specified position
- pop() - Removes the element at the specified position
- remove() - Removes the items with the specified value
- reverse() - Reverses the order of the list
- sort() - Sorts the list
