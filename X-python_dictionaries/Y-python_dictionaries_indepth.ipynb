# Python Dictionaries InDepth

## Access Dictionary Items

```python
thisdict = {
  "brand":"Ford",
  "model":"Mustang",
  "year":1964,
  "colors":["red","black","grey"]
}

print(thisdict["brand"]) # Get the value using key
print(thisdict.get("brand")) # Get the value saved in a key using get() 
```

```python
thisdict =  {
  "brand":"Ford",
  "model":"Mustang",
  "year":1964,
  "colors":["red","black","grey"]
}

print(thisdict.keys()) # Returns a list of all the keys in the dictionary
print(thisdict.values()) # Returns a list of all the values in the dictionary
print(thisdict.items()) # Returns a tuple in a list with each item in a dictionary
```

## Change Dictionary Items

```python
thisdict = {
  "brand":"Ford",
  "model":"Mustang",
  "year":1964
}
thisdict["year"] = 1999
print(thisdict)
```

```python
thisdict = {
  "brand":"Ford",
  "model":"Mustang",
  "year":1964
}
thisdict.update({"year":1999}) # update() method can be used change value in a key
print(thisdict)
```

## Add Dictionary Items

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict["color"] = "red"
print(thisdict)
thisdict.update({"engine": "v6"})
print(thisdict)
```

## Remove Dictionary Items

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.pop("model") # removes the items with the specified key name
print(thisdict)
thisdict.popitem() # removes last item (in python versions below 3.7, a random item is removed instead)
print(thisdict)
```

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict["model"] # removes the item with specified key name
print(thisdict)
thisdict.clear() # empties the dictionary
print(thisdict)
del thisdict # del can delete the dictionary completely
print(thisdict)
```

## Loop Dictionaries

```python
thisdict = {
  "name":"Dipesh",
  "age":22,
  "fav_color":"Sky Blue"
}

for x in thisdict: # gives you all the keys
  print(x, end=" ")
print()

for x in thisdict.values(): # loops through the list of values
  print(x, end=" ")
print()

for x, y in thisdict.items(): # tuple unpacking + looping through list
  print(x,y, sep="-", end=" ")
```

## Copy Dictionaries
You cannot copy a dictionary simply by typing <span style="color:red;">dict2=dict1</span>, because: dict2 will only be a reference to dict1, and changes made in dict1 will automatically also be made in dict2.

```python
thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
mydict = thisdict.copy()
print(f"thisdict: {id(thisdict)}, mydict:{id(mydict)}" )
thisdict1 = {
  "brand": "Pagani",
  "model": "Utopia",
  "year": 2022
}
mydict1 = dict(thisdict1)
print(f"thisdict: {id(thisdict1)}, mydict:{id(mydict1)}" )
```

## Nested Dictionaries

```python
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}
print(myfamily["child2"]["name"])

child1 = {
  "name" : "Emil",
  "year" : 2004
}
child2 = {
  "name" : "Tobias",
  "year" : 2007
}
child3 = {
  "name" : "Linus",
  "year" : 2011
}

myfamily1 = {
  "child1" : child1,
  "child2" : child2,
  "child3" : child3
}
print(myfamily["child1"]["name"])
```

```python
myfamily = {
  "child1" : {
    "name" : "Emil",
    "year" : 2004
  },
  "child2" : {
    "name" : "Tobias",
    "year" : 2007
  },
  "child3" : {
    "name" : "Linus",
    "year" : 2011
  }
}
for myfamily_keys, myfamily_values in myfamily.items():
  print(myfamily_keys, end=" ")
  for values in myfamily_values.values():
    print(values, end=" ")
  print()

  
    
```
