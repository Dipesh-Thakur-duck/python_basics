# Python Dictionaries
Dictionaries are used to store data values in key:value pairs.

```python
thisdict = {
  "brand":"Ford",
  "model":"Mustang",
  "year":1964
}
print(thisdict)
```

Dictionary items are ordered (as of python 3.7), changeable, and do not allow duplicates.  
Dictionary value can be accessed by using the key name.

```python
thisdict = {
  "brand":"Ford",
  "model":"Mustang",
  "year":1964
}
print(thisdict["brand"])
```

```python
print(len(thisdict)) # len() function is used to count dictionary items
thisdict = {
  "brand":"Ford",
  "model":"Mustang",
  "year":1964,
  "colors":["red", "black", "grey"] # dictionary values can be of any data type
}
print(thisdict["colors"])
```

## The dict() constructor

```python
thisdict = dict(name="Dipesh", age=22, hobby="reading books")
print(thisdict)
```
