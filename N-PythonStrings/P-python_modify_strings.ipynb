# Python - Modify Strings
Python has a set of built-in methods that you can use on strings.

## Upper Case
The <mark>upper()</mark> method returns the string in upper case.

```python
a = 'Hello, world!'
print(a.upper())
```

## Lower Case
The <mark>lower()</mark> method returns the string in lower case.

```python
a = 'Hello, world'
print(a.lower())
```

## Remove Whitespace
Whitespace is the space before and/or after the actual text, and very often you want to remove this space.  
The <mark>strip()</mark> method removes any whitespac from the beginning or the end.

```python
a = " Hello, World! "
print(a.strip())
```

## Replace String
The <mark>replace()</mark> method replaces a string with another string.

```python
a = "Hello, World!"
print(a.replace("H","J"))
```

## Split String
The <mark>split()</mark> method splits the string into substrings if it finds instance of the separater.

```python
a = "Hello, World!"
print(a.split(","))
```
