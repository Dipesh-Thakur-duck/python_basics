## Format Strings
As we know, we cannot combine strings and numbers like this: 

```python
age = 36
your_age = "I am " + age
print(your_age)
```

But we can combine strings and numbers by using f-strings or the <mark>format()</mark> method.

To specify a string as an f-string, simply put <mark>f</mark> in front of the string literal, and add curly brackets <mark>{}</mark> as placeholders for variables and other operations.

```python
age = 36
your_age = f"I am {age}"
print(your_age)
```

A placeholder can include a <i>modifier</i> to format the value.  
A modifier is included by adding a colon <mark>:</mark> followed by legal formatting type, like <mark>.2f</mark> which means fixed point number with 2 decimals:

```python
cost_price = 59
sell_price = f"Selling price: {(cost_price + cost_price * 0.1):.2f}"
print(sell_price)
```
