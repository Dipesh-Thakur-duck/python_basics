# Python Strings

## Strings
Strings in python are surrounded by either single quotation marks, or double quotation marks.  
<mark>'hello'</mark> is the same as <mark>"hello"</mark>.  
You can display a string literal (python takes the value literally as written in the code ) with the <mark>print()</mark> function.


```python
print("Hello")
print('Hello')
```

## Quotes Inside Quotes
You can use quotes inside a string, as long as they don't match the quotes surrounding the string.

```python
print("It's alright")
print("He is called 'Johnny'")
print('He is called "Johnny"')
```

## Assign String to a Variable
Assigning a string to a variable is done with the variable name followed by an equal sign and the string.

```python
a = "Hello"
print(a)
```

## Multiline Strings
You can assign a multiline string to a variable by using three double or single quotes. In result, the line breaks are inserted at the same position as in the code.

```python
a = '''My name is Dipesh. I am 22 years old.
I love my family. I am trying to be more spiritual and work hard.
'''
print(a)
```

## Strings and Arrays

Python does not have a character data type, a single character is simply a string with a length of 1. Square brackets can be used to access elements of the string

```python
a = "Hello, World!"
print(a[5])
```

Strings and arrays are different things that happen to share one behavior - indexed access.  
In Python, a string is a sequence of characters,and sequences support indexing and iteration. But that doesn't make it an array.  
<b>What makes an array an array</b> is that it stores elements of uniform type in contiguous memory, and you can freely mutate those elements. Python's <mark>list</mark> is closer to an array (and the <mark>array</mark> module gives you a true one).  
<mark>Strings break the array contract in key ways:</mark>  
- Strings are immutable
- Strings only store characters, not arbitrary elements
- Strings have <b>string-specific methods</b> that make no sense on a general array

## Looping through a String
Since strings are arrays, we can loop through the characters in a string, with a <mark>for</mark> loop.

```python
for x in "banana":
  print(x)
```

## String Length
To get the length of a string, use the <mark>len()</mark> function.

```python
a = "Aahan Thakur"
print(len(a))
```

## Check String
To check if a certain phrase or character is present in a string, we can use the keyword <mark>in</mark>.

```python
txt = "The best things in life are free!"
print("free" in txt)
```

Use it in an <mark>if</mark> statement:

```python
txt = "The best things in life are free!"
if "free" in txt:
  print("Yes, free is present.")
```

## Check if NOT

To check if a certain phrase or character is NOT present in a string, we can use the keyword <mark>not in</mark>.

```python
txt = "The best things in life are free!"
print("hello" not in txt)
```

Use it in if statement:

```python
txt = "The best things in life are free!"
if "hello" not in txt:
  print('No, hello is not present')
```

## String Concatenation
To concatenate, or combine, two strings you can use the + operator.

```python
a = "Hello"
b = "World"
c = a + " " + b
print(c)
```
