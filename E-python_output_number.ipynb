# Python Output Numbers

## Print Numbers
You can also use the <mark>print()</mark> function to display numbers. However, unlike text, we don't put numbers inside quotes.

```python
print(3)
print(358)
print(50000)
```

You can also do math inside the <mark>print()</mark> function:

```python
print(3+3)
print(2*5)
```

## Mix Text and Numbers
You can combine text and numbers in one output by separating them with a comma.

```python
print('I am', 35, 'years old.')
```
