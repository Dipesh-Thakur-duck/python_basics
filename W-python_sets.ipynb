# Python Sets
Sets are used to store multiple items in a single variable.  
A set is a collection which is <i>unordered</i>, <i>unchangeable</i>, <i>unindexed</i> and <i> do not allow duplicate values.</i>  
<span style="color:green;"><b>Note:</b> Set items are unchangeable, but you can remove items and add new items.</span>

```python
thisset = {"apple", "banana", "cherry"}
print(thisset) # Sets are unordered, so you cannot be sure in which order the items will appear.
```

## Duplicates Not Allowed
Sets cannot have two items with the same value.

```python
thisset = {"apple", "banana", "cherry", "apple"}
print(thisset)
```

```python
set1 = {"apple", 1, True, "cherry"} # The values True and 1 are considered the same value in sets, and are treated as duplicates. Same goes for False and 0.
print(set1)
```

## Get the Length of a Set

```python
set2 = {"apple", "banana", "cherry"}
print(len(set2))
```

## The set() Constructor
It is also possible to use the set() constructor to make a set.

```python
set3 = set(("apple", "banana", "cherry"))
print(set3)
```
