# Python Operators
Operators are used to perform operations on variables and values.  

## Arithmetic Operators
Arithmetic operators are used with numeric values to perform common mathematical operations.  
They are:
- '+' -> Addition
- '-' -> Subtraction
- '*' -> Multiplication
- '/' -> Division
- '%' -> Modulus (gives the remainder)
- '**' -> Exponentiation
- '//' -> Floor Division (rounds DOWN to nearest integer)

### Division in Python
Python has two division operators:  
- '/' - Division (returns a float)
- '//' - Floor Division (returns an integer)

```python
x = 12
y = 5

print(x / y)

x = 12
y = 5

print(x // y) # rounds DOWN to the nearest integer
```

```python
x = 5
y = 2

print(x % y) # gives the remainder
```

## Assignment Operators in Python

### Basic Assignment
Assigns a value to a variable.

```python
x = 10
name = "Dipesh"
```

### Augmented Assignment Operators
These perfrom an operation and assign the result in one step.  

```python
x = 10
x += 5   # x = x + 5, 15
x -= 3   # x = x - 3, 12
x *= 2   # x = x * 2, 24
x /= 4   # x = x / 4, 6.0   (always returns float)
x //= 2  # x = x // 2, 3.0   (floor division)
x %= 2   # x = x % 2, 1.0   (remainder)
x **= 3  # x = x ** 3, 1.0   (power)
```

### Bitwise Assignment Operators
Used when working with binary/bits.

```python
x = 6        # binary: 110
x &= 3       # 110 & 011 = 010 → 2
x |= 5       # 010 | 101 = 111 → 7
x ^= 3       # 111 ^ 011 = 100 → 4 (exclusive OR)
x >>= 1      # 100 >> 1  = 010 → 2  (divide by 2)
x <<= 2      # 010 << 2  = 1000 → 8 (multiply by 4)
```

### Walrus Operator - := (Python 3.8+)
Assigns a value inside an expression (assign and use at the same time).

```python
# Without walrus
data = input("Enter: ")
if data:
    print(data)

# With walrus — cleaner
if data := input("Enter: "):
    print(data)
```

## Ternary Operator
The ternary operator allows you to assign one value if a condition is true, and another if it is false.  
It is not an actual operator, it is a conditional expression or a shorthand if statement.

```python
num = 6

x = "WEEKEND!" if num > 5 else "Workday"

print(x)
```

The ternary operator can be used instead of <mark>elif</mark> in longer if statements.

```python
num = 6

x = "Fri" if num == 5 else "Sat" if num == 6 else "Sun" if num == 7 else "weekday"

print(x)
```

## Comparison Operators  
Comparison operators are used to compare two values.

```python
x = 5
y = 3

print(x == y) # Equal
print(x != y) # Not Equal
print(x > y) # Greater than
print(x < y) # Less than
print(x >= y) # Greater than or equal to
print(x <= y) # Less than or equal to
```

### Chaining Comparison Operators
Python allows you to chain comparison operators.

```python
x = 5

print(1 < x < 10)

print(1 < x and x < 10)
```

## Logical operators
They are used to combine or modify conditions - they return True or False based on the logic.

```python
x = 5
y = 10

if x == 5 and y == 10: # returns True if both statements are true
  print("It's a match")
if x > 3 or y < 9: # returns True if at least one statement is true
  print("At least one condition is true")
if not x > 6:
  print('X is indeed less than 6')
```

## Python Identity Operators
Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location. 

```python
x = ["apple", "banana"]
y = ["apple", "banana"]
z = x

print(x is z) # points to the same memory location
print(x is y) # points to different memory location
print(x == y) # are the values equal??
```

## Membership Operators
They are used to check if a value exists inside a sequence (list, string, tuple, dict, set).

```python
fruits = ["apple", "banana", "mango"]

print("apple" in fruits)   # True
print("grape" in fruits)   # False

user = {"name": "Dipesh", "age": 20}

print("name" in user)       # True  (checks keys)
print("Dipesh" in user)     # False (values not checked)

# To check values:
print("Dipesh" in user.values())   # True
```

```python
print("grape" not in fruits)   # True
print("apple" not in fruits)   # False
```

## Bitwise Operators 
Bitwise operators work on integers at the binary level.
- AND (&): 1 only if both bits are 1
- OR (|): 1 if either bit is 1
- XOR (^): 1 if bits are different
- NOT (~): Flips all bits
- Left Shift: (<<): Shifts bits left (multiply by 2^n)
- Right Shift: (>>): Shifts bits right (divide by 2^n)

```python
a = 0b1010  # 10
b = 0b1100  # 12

print(a & b)   # 8  → 0b1000  (AND)
print(a | b)   # 14 → 0b1110  (OR)
print(a ^ b)   # 6  → 0b0110  (XOR)
print(~a)      # -11           (NOT — flips bits, result = -(n+1))
print(a << 1)  # 20 → 0b10100 (left shift by 1)
print(a >> 1)  # 5  → 0b0101  (right shift by 1)
```
