# Python Output / Print

## Print Text
<mark>print()</mark> function is used to display text or output values

```python
print('Hello World!')
```

## Double "Quotes
Text in Python must be inside quotes. You can use either <span style="color:red;">"<span> or <span style="color:red;">'<span> single quotes

## Print Without a New Line
By default, the <mark>print()</mark> function ends with a new line. If you want to print multiple words on the same line, you can use the <span style="color:red;">end</span> parameter:

```python
print('Hello World!', end=' ')
print("I will print on the same line.")
```

Note that we add a space after the <span style="color:red;">end=" "</span> for better readability.
