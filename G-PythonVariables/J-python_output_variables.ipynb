# Python - Output Variables

## Output Variables
The print <mark>print()</mark> function is often used to output variables.

```python
x = "Python is awesome"
print(x)
```

In the <mark>print()</mark> function, you can output multiple variables, separated by a comma.

```python
x = "Python"
y = "is"
z = "awesome"
print(x, y, z)
```

You can also use the <mark>+</mark> operator to output multiple variables.

```python
x = "Python "
y = "is "
z = "awesome"
print(x+y+z)
```

Notice the space character after <mark>"Python"</mark> and <mark>"is"</mark>, without them the result would be "Pythonisawesome".

For numbers, the <mark>+</mark> character works as a mathematical operator.

```python
x = 5
y = 10
print(x+y)
```

In the <mark>print()</mark> function, when you try to combine a string and a number with the <mark>+</mark> operator, Python will give you an error.  
The best way to output multiple variables in the <mark>print()</mark> function is to separate them with commas, which even support different data types.

```python
x = 5
y = "John"
print(x, y)
```
