# Python Booleans
Booleans represent one of the two values: <mark>True</mark> or <mark>False</mark>

In programming you often need to know if the statement is True or False.  
You can evaluate any expression in Python, and get one of two answers, True or False.

```python
print(10 > 9)
a = 10
b = 9
if a > 9:
  print("10 is greater than 9")
```

## bool() function
The <mark>bool()</mark> function allows you to evaluate any value, and give you <mark>True</mark> or <mark>False</mark> in return.

```python
print(bool('Hello'))
print(bool(56))
```

## Most Values are True
Almost any value is evaluated to <mark>True</mark> if it has some sort of content.  
Any string is True, except empty strings.  
Any number is True, except 0.  
Any list, tuple, set, and dictionary are True, except any empty ones.  
## Some Values are False
The value <mark>None</mark> evaluates to False.  

Python has many built-in functions that return a boolean value, like the <mark>isinstance()</mark> function, which can be used to determine if an object is of a certain data type.

```python
x = 200
print(isinstance(x, int))
```
