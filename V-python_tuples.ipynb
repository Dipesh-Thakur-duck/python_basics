# Python Tuples
Tuples are used to store multiple items in a single variable, created with round brackets.  
Tuple items are ordered, unchangeable, and allow duplicate values.  
Ordered: Items have a defined order, and that order will not change.  
Unchangeable: It means we cannot change, add or remove items after the tuple has been created.

```python
thistuple = ("apple", "banana", "cherry")
print(type(thistuple))  # <class 'tuple'>
```

## Tuple Length
Use len() function

```python
thistuple = ("apple", "banana", "cherry")
print(len(thistuple))
```

## Create Tuple with One item
To create a tuple with only one item, you have to add comma after the item.

```python
thistuple = ("apple",)
print(type(thistuple))

#NOT a tuple
thistuple = ("apple")
print(type(thistuple))
```

## The tuple() constructor
Using tuple() constructor to make a tuple

```python
thistuple = tuple(("apple", "banana", "cherry")) # note the double round-brackets
print(thistuple)
```

## Accessing Items

```python
thistuple = ("apple", "banana", "cherry")

thistuple[0]    # "apple"   (first)
thistuple[-1]   # "cherry"  (last)
thistuple[1:3]  # ("banana", "cherry")
thistuple[:2]   # ("apple", "banana")
thistuple[1:]   # ("banana", "cherry")
```

### Check existence

```python
if "apple" in thistuple:
  print("Found!")

thistuple = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5)

x = thistuple.count(5)

print(x) 
```

## Tuples are unchangeable - What That Really Means
You cannot do <span style="color:red;">thistuple[0] = "mango"</span> - this raises a <mark>TypeError</mark>.  
But <span style="color:red;">tup += (1, )</span> does work. It's not mutating the tuple. Under the hood it does: 

```python
tup = (1, 2, 3)
tup = tup + (4, ) # creates a BRAND NEW tuple, reassigns the variable
```

```python
tup = (1, 2, 3)
ref = tup
print(id(tup))   # e.g. 140234567

tup += (4,)
print(id(tup))   # DIFFERENT — it's a new object
print(ref)       # (1, 2, 3) — original still unchanged
```

With a list, <mark>ref</mark> would also see the change becuase lists mutate in place. With tuples, the original is never touched - a new object is always created.

## Workaround to "Change" a Tuple 
Since tuples are immutable, the standard workaround is to convert to a list, modify, then convert back: 

```python
x = ("apple", "banana", "cherry")
y = list(x)       # convert to list
y[1] = "mango"    # modify
x = tuple(y)      # convert back
print(x)          # ("apple", "mango", "cherry")
```

Or just concatenate with <mark>+=</mark> (creates a new tuple).

## Unpacking Tuples
Assigning tuple items directly to variables: 

```python
fruits = ("apple", "banana", "cherry")
(a, b, c) = fruits   # parentheses optional
print(a)  # apple
print(b)  # banana
print(c)  # cherry
```

Using * to collect the rest: 

```python
fruits = ("apple", "banana", "cherry", "mango", "kiwi")

(a, *b, c) = fruits
# a = "apple"
# b = ["banana", "cherry", "mango"]  ← becomes a list
# c = "kiwi"
```

Only one * variable allowed per unpack statement.

## Looping

```python
thistuple = ("apple", "banana", "cherry")

# for loop
for x in thistuple:
    print(x)

# loop by index
for i in range(len(thistuple)):
    print(thistuple[i])

# while loop
i = 0
while i < len(thistuple):
    print(thistuple[i])
    i += 1
```

## Joining Tuples and Multiply Tuples

```python
tuple1 = ("a", "b", "c")
tuple2 = (1, 2, 3)

tuple3 = tuple1 + tuple2       # ("a", "b", "c", 1, 2, 3)
mytuple = ("apple",) * 3       # ("apple", "apple", "apple")
```

## Tuples Methods
Only 2 built-in methods since tuples can't be modified: 
- count(x): Returns number of times x appears
- index(x): Returns index of first occurrence of x

```python
thistuple = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5)
print(thistuple.count(5))   # 2
print(thistuple.index(8))   # 3
```

## Why Use Tuples Over Lists?  
- Faster than lists for iteration and access
- Write protection : use when data shouldn't change (e.g. coordinates, RGB values, DB records)
- Can be used as dictionary keys (lists cannot, since they're mutable)
- Slightly less memory usage
